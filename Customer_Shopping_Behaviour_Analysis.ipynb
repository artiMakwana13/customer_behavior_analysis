```python
# df.groupby('Category'): splits the DataFrame into groups by category
# df.groupby('Category')['Review Rating']: From each group, we take only the Review Rating column.
#.transform(...): transform() function applies an operation to each group, and returns a Series of same length
# lambda x: x.fillna(x.median()): x represents the Review Rating within one group (category)
# x.fillna(x.median()): replaces all missing (NaN) values in that group with the median of that group.

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
# create a column for age group

labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly': 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
